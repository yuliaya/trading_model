```python
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
import pickle
```

```python
brands = ['Chanel', 'Louis Vuitton', 'Fendi', 'Saint Laurent', 'Gucci']
data_dir = './data/demand_prediction/'
ts_dict = {}
```

```python
for brand in brands:
    ts_dict[brand] = {}
    for file in [f for f in os.listdir(data_dir) if re.search('%s_search'%brand, f)]:
        search_df = pd.read_csv(data_dir+'%s' %file)
        date_col = search_df.columns[0]
        stat_col = search_df.columns[1]
        search_df[date_col] = pd.to_datetime(search_df[date_col]).dt.date
        search_df[stat_col] = search_df[stat_col].astype(int)
        if brand == 'Chanel' and file == 'Chanel_search_chanel.csv':
            chanel_df = search_df.copy()
        search_df[stat_col].plot()
        series = search_df[stat_col]
        result = seasonal_decompose(series, model='additive', freq=7)
        search_df[stat_col] = result.observed - result.seasonal
        if brand == 'Chanel' and file == 'Chanel_search_chanel.csv':
            chanel_df['refined'] = search_df[stat_col]
        search_df[stat_col].plot()
        print(brand, file)
        plt.show()
        search_dict = dict(zip(search_df[date_col],
            [int(item) for item in search_df[stat_col]]))
        keyword = 'search_' + file.split('_search_')[1][:-4]
        ts_dict[brand][keyword] = search_dict
```

```python
with open(data_dir+'search_trends.pkl', 'wb') as f:
    pickle.dump(ts_dict, f)
```
